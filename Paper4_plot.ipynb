```python
import numpy
import matplotlib.pyplot as plt
import pandas
import math
```

```python
df_hour = pandas.read_csv('Paper4-plot-hour.csv', usecols=[0,1,2,3], engine='python')
df_halfhour = pandas.read_csv('Paper4-plot-halfhour.csv', usecols=[0,1,2,3], engine='python')
df_15min = pandas.read_csv('Paper4-plot-15min.csv', usecols=[0,1,2,3], engine='python')
df_10min = pandas.read_csv('Paper4-plot-10min.csv', usecols=[0,1,2,3], engine='python')
```

```python
plt.subplot(2, 2, 1)
plt.plot(df_hour['GT'], 'ro', df_hour['LSTM'], '-', df_hour['MLP'], '--', df_hour['SVR'] ,':', markersize=3)
#plt.legend(['GT', 'LSTM', 'FFNN', 'SVR'], loc='best')
#plt.xticks(numpy.arange(0, 81, 20))
plt.ylabel('Flow (m3/s)')
plt.title('(a) One hour frequency')
#plt.yticks(numpy.arange(0, 1.41, 0.2))

plt.subplot(2, 2, 2)
plt.plot(df_halfhour['GT'], 'ro', df_halfhour['LSTM'], '-', df_halfhour['FFNN'], '--', df_halfhour['SVR'] ,':', markersize=3)
#plt.legend(['GT', 'LSTM', 'FFNN', 'SVR'], loc='best')
#plt.xticks(numpy.arange(0, 81, 20))
plt.title('(b) Half hour frequency')
#plt.yticks(numpy.arange(0, 1.41, 0.2))

plt.subplot(2, 2, 3)
plt.plot(df_15min['GT'], 'ro', df_15min['LSTM'], '-', df_15min['FFNN'], '--', df_15min['SVR'] ,':', markersize=2)
#plt.legend(['GT', 'LSTM', 'FFNN', 'SVR'], loc='best')
#plt.xticks(numpy.arange(0, 81, 20))
plt.xlabel('Time steps')
plt.ylabel('Flow (m3/s)')
plt.title('(c) 15 Minutes frequency')
#plt.yticks(numpy.arange(0, 1.41, 0.2))

plt.subplot(2, 2, 4)
plt.plot(df_10min['GT'], 'ro', df_10min['LSTM'], '-', df_10min['FFNN'], '--', df_10min['SVR'] ,':', markersize=2)
plt.legend(['GT', 'LSTM', 'FFNN', 'SVR'], loc='best')
#plt.xticks(numpy.arange(0, 81, 20))
plt.xlabel('Time steps')
plt.title('(d) 10 Minutes frequency')
#plt.yticks(numpy.arange(0, 1.41, 0.2))

plt.subplots_adjust(bottom=0.25, right=1.5, top=1.7)
plt.savefig("Paper4-NN.pdf", bbox_inches='tight', dpi=1500)
plt.savefig("Paper4-NN.png", bbox_inches='tight', dpi=1000)
plt.show()
```
